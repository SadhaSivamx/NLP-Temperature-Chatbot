```python
import random
import json
import numpy as np
import pickle
import nltk
from nltk.stem import WordNetLemmatizer
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation,Dropout
from tensorflow.keras.optimizers import SGD
import nltk
nltk.download('punkt')
from nltk import word_tokenize,sent_tokenize
```

```python
temperature=32
humidity=72
lemmatizer=WordNetLemmatizer()
intents=json.loads(open("intents.json").read())
```

```python
words=[]
classes=[]
documents=[]
ignored=['?','!','.',',',"'s"]
```

```python
for _ in intents['intents']:
  for text in _['patterns']:
    wordlist=nltk.word_tokenize(text)
    words.extend(wordlist)
    documents.append((wordlist,_['tag']))
    if _['tag'] not in classes:
      classes.append(_['tag'])
print(documents)
print(words)
```

```python
doccy=[]
for x in words:
  if x not in ignored:
    doccy.append(lemmatizer.lemmatize(x))
doccy=sorted(set(doccy))
words=doccy
words
```

```python
classes=sorted(set(classes))
pickle.dump(words,open("words.pkl","wb"))
pickle.dump(classes,open("classes.pkl","wb"))
training=[]
output_empty=[0]*len(classes)
```

```python
for document in documents:
  bag=[]
  word_patterns=document[0]
  word_patterns=[lemmatizer.lemmatize(word.lower()) for word in word_patterns]
  for _ in words:
    bag.append(1) if _ in word_patterns else bag.append(0)
  output_row=list(output_empty)
  output_row[classes.index(document[1])]=1
  training.append([bag,output_row])
random.shuffle(training)
training=np.array(training)
training
```

```python
train_x=list(training[:,0])
train_y=list(training[:,1])
model=Sequential()
model.add(Dense(128,input_shape=(len(train_x[0]),),activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(64,activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(len(train_y[0]),activation="softmax"))
dgd=SGD(lr=0.01,momentum=0.9,nesterov=True)
model.compile(loss='categorical_crossentropy',optimizer=dgd,metrics=['accuracy'])
hist=model.fit(np.array(train_x),np.array(train_y),epochs=200,batch_size=5,verbose=1)
model.save('chatbot.h5',hist)
print('done')
```

```python
temp='32*c'
humi='72%'
def cleanup(sentenze):
  sw=nltk.word_tokenize(sentenze)
  sw=[lemmatizer.lemmatize(word) for word in sw]
  return sw
def bag_of_words(sentenze):
  sw=cleanup(sentenze)
  bag=[0]*len(words)
  for w in sw:
    for i,j in enumerate(words):
      if j==w:
        bag[i]=1
  return np.array(bag)
def predictclass(sent):
  bow=bag_of_words(sent)
  res=model.predict(np.array([bow]))[0]
  error=0.25
  result=[[i,r] for i,r in enumerate(res) if r>error]
  result.sort(key=lambda x: x[1],reverse=True)
  retur=[]
  for _ in result:
    retur.append({'intent':classes[_[0]],'probability':str(_[1])})
  return retur
def getres(intents_list,intents_json):
  tag=intents_list[0]['intent']
  listof=intents_json['intents']
  for i in listof:
    if i['tag']==tag:
      result=random.choice(i['responses'])
      break
  return result
print("Go! Bot is running!")
while True:
  sentence=input("Message : ")
  ints=predictclass(sentence)
  res=getres(ints,intents)
  try:
    print(res.format(temp,humi))
  except:
    print(res)
```
